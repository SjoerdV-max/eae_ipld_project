```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ex 2.1: Load the data using Pandas read_csv
data_path = "../data/netflix_titles.csv"

movies_df = pd.read_csv(data_path, index_col="show_id")  # Load CSV and set 'show_id' as index

movies_df  # Display the DataFrame
```

```python
# Ex 2.2: Find the min and max release years

min_year = movies_df["release_year"].min()  # Get the minimum release year
max_year = movies_df["release_year"].max()  # Get the maximum release year

print(f"Min year: {min_year}, Max year: {max_year}")
```

```python
# Ex 2.3: Count missing values in the 'director' column

num_missing_directors = movies_df["director"].isna().sum()  # Count NaN values

print(f"Number of missing directors: {num_missing_directors}")
```

```python
# Ex 2.4: Count the number of unique countries

# Step 1: Fill NaN values with "Unknown"
movies_df["country"] = movies_df["country"].fillna("Unknown")

# Step 2: Create a list of all individual countries
all_countries = ", ".join(movies_df["country"]).split(", ")  # Join all country entries and split by ", "

# Step 3: Get the unique countries
n_countries = len(set(all_countries))  # Convert to set to get unique values

print(f"There are {n_countries} different countries in the data")
```

```python
# Ex 2.5: Compute the average title length

# Step 1: Create a new column with title lengths
movies_df["title_length"] = movies_df["title"].apply(lambda x: len(x))

# Step 2: Compute the average title length
avg_title_length = movies_df["title_length"].mean()

print(f"The average title length is {avg_title_length:.2f} characters")
```

```python
# Ex 2.7: Average duration of movies per year

# Step 1: Filter only movies (not TV shows)
movies_only = movies_df[movies_df["type"] == "Movie"]

# Step 2: Extract duration in minutes as an integer
movies_only["duration_minutes"] = movies_only["duration"].str.extract("(\d+)").astype(float)

# Step 3: Compute the average duration per year
movies_avg_duration_per_year = movies_only.groupby("release_year")["duration_minutes"].mean()

# Step 4: Plot the line chart
fig = plt.figure(figsize=(9, 6))
plt.plot(movies_avg_duration_per_year.index, movies_avg_duration_per_year, marker="o", linestyle="-")
plt.xlabel("Year")
plt.ylabel("Average Duration (minutes)")
plt.title("Average Duration of Movies Across Years")

plt.show()
```
